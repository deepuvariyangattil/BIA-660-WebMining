GIVING NAMES TO EACH TOPIC- I HAVE TAKEN TWO WORDS WITH HIGHEST WEIGHT AND CONCATENATED IT WITH '&'

```python
import re
```

```python
lda = gensim.models.ldamodel.LdaModel(corpus=mm, id2word=id2word, num_topics=20, update_every=1, passes=1)
```

```python
c=[]
for topic in lda.print_topics(20):
    a=list(topic)[1].split('+')
    for b in a:
        c.append((re.findall("[a-z]+",b)))
        
    
    topic_name=str(c[0])+"&"+str(c[1])
    #print(type(str(c[0])))
    s=list(topic)
    topic_name=topic_name.replace("[","")
    topic_name=topic_name.replace("]","")
    s[0]=topic_name
    print(s)
    c=[]
    
    print("\n")
```

TESTING WIKI RANDOM PAGES

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
```

```python
from stop_words import get_stop_words

# create English stop words list
en_stop = get_stop_words('en')
```

```python
from nltk.stem.porter import PorterStemmer

# Create p_stemmer of class PorterStemmer
p_stemmer = PorterStemmer()
```

```python
control=[1,2,3]
for i in control:
    print("web page"+str(i))
    temp="wikirandom"+str(i)
    file=open(temp+".txt",'r',encoding="utf-8")
    doc=file.read()
    raw = doc.lower()
    tokens = tokenizer.tokenize(raw)
    stopped_tokens = [i for i in tokens if not i in en_stop]
    texts = [p_stemmer.stem(i) for i in stopped_tokens]
    dataset = [d.split() for d in texts]
    dictionary = gensim.corpora.Dictionary(dataset)
    bow_vector = dictionary.doc2bow(tokens)
    for index, score in sorted(lda[bow_vector], key=lambda tup: -1*tup[1]):
        
        print("Score: {}\t Topic: {}".format(score, lda.print_topic(index, 10)))
```
