```python
import networkx as nx
```

```python
edges=[]
```

```python
f=open("web-edu.mtx","r",encoding="utf-8")
for line in f:
    #print (line)
    
    line=line.replace('\n','')
    
    
    edge1,edge2=line.split(" ")
    newedge=(edge1,edge2)
    print(newedge)
    edges.append(newedge)
    
```

```python
print(edges)
```

```python
G=nx.Graph()
```

```python
G.add_edges_from(edges)
```

```python
nx.draw(G)
```

```python
G.number_of_nodes()
```

```python
G.number_of_edges()
```

```python
G.degree('2818')
```

```python
G.degree('342')
```

```python
output=nx.pagerank(G)
```

```python
sorted_by_value = sorted(output.items(), key=lambda kv: kv[1],reverse=True)
```

```python
sorted_by_value
```
